3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
# Функция для обработки опыта работы
def experience(x): 
    if pd.isnull(x) or x == 'Не указано': # Если данные отсутствуют или "Не указано"
        return None
    
    # Разделяем строку по пробелам и берем первые 5 элементов
    x = x.split(' ')[:5]
    
    if 'месяц' in x or 'месяца' in x or 'месяцев' in x:
        return int(x[2]) # Возвращаем количество месяцев
    elif 'год' in x or 'лет' in x or 'года' in x: # Проверяем, присутствуют ли слова
        # Если в конце списка пробел, то там остались только года
        if x[-1] == '':
            return int(x[2]) * 12 # Преобразуем количество лет в месяцы
        else:
            return int(x[2]) * 12 + int(x[-1]) # Месяцы и годы вместе
            
            
    
    
data['Опыт работы'] = data['Опыт работы'].apply(experience) # Применяем функцию

# Задание для площадки
display(data['Опыт работы'].median()) # Находим медиану


```

```python
#ваш код здесь
# Список городов миллиоников
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород',
                  'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
                  'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

# Список двух основных городов
normal_cities = ['Москва', 'Санкт-Петербург']

# Функция для распределения по городам
def cities(x):
    if x[0] in normal_cities: # Проверка на наличие города в списке основных городов
        return x[0]
    elif x[0] in million_cities: # Проверка на наличие в списке городов миллиоников
        return "Город-миллионник"
    else: # Если не подходит ни под какую категорию - отправляем в 'другое'
        return 'Другие'

# Функция для распределения по готовности переехать    
def ready_to_move(x):
    if 'не готов к переезду' in x or 'не готова к переезду' in x: # Проверка на готовность переехать
        return False    
    else:
        return True
    
# Функция для проверки на готовность командировок
def ready_to_trip(x):
    if 'командировкам' in x: # Проверка на наличие информации о командировках
        if 'не готов к командировкам' in x or'не готова к командировкам' in x: # Проверка на готовность к командировкам
            return False
        else: 
            
            return True
    else:
        return False

temp_data = data['Город, переезд, командировки'].apply(lambda x: x.split(' , ')) # Обрабатываем данные в список для возможности обратиться по индексу
data['Город'] = temp_data.apply(cities) # Применяю функции для городов

metro_names = temp_data.apply(lambda x: x[1].strip() if 'м.' in x[1] else None) # Находим метро в данных и убираем пробел в начале строк
metro_list = metro_names.unique() # Создаем список уникальных названий метро
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(ready_to_move)
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(ready_to_trip)
data = data.drop(columns=['Город, переезд, командировки'])

# Код для площадки
display(data['Город'].value_counts(normalize=True) * 100) # Процент живущих в по признакам
ready_both = data[data['Готовность к переезду'] & data['Готовность к командировкам']] # Фильтруем только тех, кто готов к обоим
percent = (len(ready_both) / len(data)) * 100 # Высчитываем процент готовных к переезду и командировкам
display(round(percent))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# Преображение категории "Занятость"
employments = ['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']

for employment in employments:
    data[employment] = data['Занятость'].apply(lambda x: employment in x)

data = data.drop('Занятость', axis=1)


# Преображение категории "График"
schedules = ['гибкий график', 'полный день', 'сменный график', 'вахтовый метод', 'удаленная работа']

for schedule in schedules:
    data[schedule] = data['График'].apply(lambda x: schedule in x)

data = data.drop('График', axis=1)


# Код для площадки
project_and_volunteering = data[data['проектная работа'] & data['волонтерство']]
print(project_and_volunteering.shape[0])
rotshift_and_flexschedule = data[data['вахтовый метод'] & data['гибкий график']]
print(rotshift_and_flexschedule.shape[0])
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
# Приводим к формату datetime
data['date'] = pd.to_datetime(data['Обновление резюме'], dayfirst=True, errors='coerce').dt.date
# Читаем df с данными ISO
exchange_data = pd.read_csv('D:\IDE\My_commits\HeadHunterResumeAnalysis\ExchangeRates.csv')
# Приводим дату в формат datetime
exchange_data['date'] = pd.to_datetime(exchange_data['date'], format='%d/%m/%y', dayfirst=True).dt.date
data['Обновление резюме'] = data['date']
# Избавляемся от ненужных столбцов
exchange_data = exchange_data.drop(columns=['per', 'time', 'vol'])

# Словарь - (название валюты: формат в ISO)
iso_dict = {'грн.':'UAH', 'USD':'USD', 'EUR':'EUR', 'бел.руб.':'BYN', 'KGS':'KGS', 'сум':'UZS', 'AZN':'AZN', 'KZT':'KZT', 'руб.':'руб'}
#  Переводим наименование валюты в стандарт ISO
data['currency'] = data['ЗП'].apply(lambda x: iso_dict[x.split(' ')[1]])
# Отдельно создаем столбец с зп
data['ЗП'] = data['ЗП'].apply(lambda x: x.split(' ')[0])

# Объединяем dataframe 
data = data.merge(exchange_data, on=['date', 'currency'], how='left')
# Заполняю пропуски, в данном случае для рублей
data['proportion'] = data['proportion'].fillna(1)
data['close'] = data['close'].fillna(1)
# Высчитываю зарплату по формуле: умножить заработную плату на присоединённый курс валюты и разделить на пропорцию
data['ЗП (руб)'] = (data['ЗП'].astype(int) * data['close'].astype(float)) / data['proportion'].astype(int)
# Удаляю более ненужные столбцы
data = data.drop(columns=['ЗП', 'close', 'proportion', 'currency', 'date'])

# Код для площадки
display(data['ЗП (руб)'].median())

```
